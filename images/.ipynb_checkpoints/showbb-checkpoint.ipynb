```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg
import matplotlib.lines as mlines
import cv2
from ast import literal_eval
import numpy as np
import os
from IPython.display import Image, display
from ast import literal_eval
```

```python
filelist=os.listdir('Testgame_22_12_17_Hulk')
for fichier in filelist[:]: # filelist[:] makes a copy of filelist.
    if not(fichier.endswith(".png")):
        filelist.remove(fichier)

rows={}
data = [[item for item in line.rstrip('\r\n').split('|')] 
         for line in open('Testgame_22_12_17_Hulk/base.txt')]
for i in data:
    i.remove('ball')
    i.remove('')
    rows[i[0]]=i[1]
```

```python
for i in filelist:
    value = rows[i]
    dic = literal_eval(value)
    print(dic)
#   display(Image('Testgame_22_12_17_Hulk/' + i))
```

```python
for i in filelist:
    if i == '26_01_2018__19_38_28_0488_upper.png':
        value = rows["26_01_2018__19_38_28_0488_upper.png"]
        dic = literal_eval(value)

        # data points
        x1 = int(dic['x1'])
        x2 = int(dic['x2'])
        y1 = int(dic['y1'])
        y2 = int(dic['y2'])

        img = cv2.imread('Testgame_22_12_17_Hulk/26_01_2018__19_38_28_0488_upper.png')
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        pt1 = (x1, y1)
        pt2 = (x2, y2)
        cv2.rectangle(img, pt1, pt2, (255,0,0), lineType=8, shift=0)
        cv2.imshow('hoi', img)
        cv2.waitKey(50)
```
